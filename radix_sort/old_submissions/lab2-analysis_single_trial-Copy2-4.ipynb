```python
files = glob("m2_cali/*.cali")

print(f"Total number of files found: {len(files)}")


valid_files = []  # List to keep track of valid files
skipped_files = []  # List to keep track of files that caused errors
processed_count = 0  # Counter for successfully processed files

# First, collect all valid files
for file in files:
    try:
        print(f"Processing file: {file}")
        # Try to process the file to check if it's valid
        tk = th.Thicket.from_caliperreader([file])
        print(f"Successfully processed: {file}")
        valid_files.append(file)  # Append to valid files if successful
        processed_count += 1
    except KeyError as e:
        print(f"Error processing file: {file}, skipping...")
        skipped_files.append(file)  # Add the problematic file to the skipped list

# If there are valid files, load them into a single Thicket object
if valid_files:
    print("Combining all valid files into a single Thicket object.")
    combined_tk = th.Thicket.from_caliperreader(valid_files)
else:
    print("No valid files to combine.")

# After the loop, print out a summary
print(f"\nSummary:")
print(f"Successfully processed {processed_count} files.")
if skipped_files:
    print(f"{len(skipped_files)} files were skipped due to errors:")
    for skipped_file in skipped_files:
        print(skipped_file)
else:
    print("No files were skipped.")
```

```python
def plot_pandas(df, title, xlabel, ylabel):
    plot = df.plot(
        kind="line", 
        marker='o', 
        title=title, 
        xlabel=xlabel, 
        ylabel=ylabel, 
        figsize=(15,7)
    )
    return plot

# For each input size (strong scaling)
for input_size in matrix_sizes:
    df_combined = pd.DataFrame()
    print(f"Processing input size: {input_size}")
    
    # For each input type (lines on the same plot)
    for input_type in sort_type:
        print(f"Processing input type: {input_type}")
        
        # Filter the dataframe by input_size and input_type columns
        df_filtered = combined_tk.dataframe[
            (combined_tk.dataframe['input_size'] == input_size) & 
            (combined_tk.dataframe['input_type'] == input_type)
        ]
        
        print(df_filtered)
        
        # Combine processes (x-axis) and Avg time/rank (y-axis)
        if not df_filtered.empty:
            df_combined[input_type] = df_filtered.groupby("num_tasks")['Avg time/rank'].mean()

    # Plot the results if there is data to plot
    if not df_combined.empty:
        plot_pandas(
            df=df_combined,
            title=f"Strong Scaling - Input Size: {input_size}",
            xlabel="Processes",
            ylabel="Time (seconds)"
        )
        plt.savefig(f"strong_scaling_input_size_{input_size}.png")
```

```python
# Weak scaling plots for each input_type
for input_type in sort_type:
    df_combined = pd.DataFrame()
    print(f"Processing weak scaling plot for input type: {input_type}")
    
    # Loop through all input sizes
    for input_size in matrix_sizes:
        print(f"Processing input size: {input_size}")
        
        df_time_vs_process = pd.DataFrame()  # Temp DataFrame to hold process vs time
        
        # For each process, get the corresponding time for the input_size and input_type
        for process in processes:
            # Filter the dataframe by input_size, input_type, and num_tasks
            df_filtered = combined_tk.dataframe[
                (combined_tk.dataframe['input_size'] == input_size) & 
                (combined_tk.dataframe['input_type'] == input_type) & 
                (combined_tk.dataframe['num_tasks'] == process)
            ]
            
            if not df_filtered.empty:
                # Get the average time for the current process count
                avg_time = df_filtered['Avg time/rank'].mean()
                
                # Append to the temporary DataFrame for this input_size
                df_time_vs_process.loc[process, 'Avg time/rank'] = avg_time
        
        # If there is data for this input size, add it as a line in df_combined
        if not df_time_vs_process.empty:
            df_combined[input_size] = df_time_vs_process['Avg time/rank']

    # Plot the results if there is data to plot
    if not df_combined.empty:
        plot_pandas(
            df=df_combined,
            title=f"Weak Scaling - Input Type: {input_type}",
            xlabel="Processes (num_tasks)",
            ylabel="Time (seconds)"
        )
        plt.savefig(f"weak_scaling_input_type_{input_type}.png")
        # plt.close()  # Close the plot to avoid overlap in subsequent plots
```

```python
# Strong scaling speedup plot for each input type
for input_type in sort_type:
    df_combined = pd.DataFrame()
    print(f"Processing strong scaling speedup plot for input type: {input_type}")
    
    # For each input size (lines on the same plot)
    for input_size in matrix_sizes:
        print(f"Processing input size: {input_size}")
        
        # Filter the dataframe by input_size and input_type
        df_filtered = combined_tk.dataframe[
            (combined_tk.dataframe['input_size'] == input_size) & 
            (combined_tk.dataframe['input_type'] == input_type)
        ]
        
        if not df_filtered.empty:
            # Group by number of tasks (processes) and calculate the average time for each
            avg_time_per_task = df_filtered.groupby("num_tasks")['Avg time/rank'].mean()
            
            # Find the smallest number of processes (this will be our baseline for speedup)
            smallest_num_tasks = avg_time_per_task.index.min()  # Get the smallest number of tasks
            baseline_time = avg_time_per_task[smallest_num_tasks]  # T1 (the baseline time)
            
            # Calculate speedup as T1 / Tn for each number of tasks
            speedup = baseline_time / avg_time_per_task
            
            # Add the speedup for this input_size to the combined dataframe
            df_combined[input_size] = speedup
    
    # Plot the speedup results if there is data to plot
    if not df_combined.empty:
        plot_pandas(
            df=df_combined,
            title=f"Strong Scaling Speedup - Input Type: {input_type}",
            xlabel="Processes (num_tasks)",
            ylabel="Speedup"
        )
        plt.savefig(f"strong_scaling_speedup_input_type_{input_type}.png")
        # plt.close()  # Close the plot to avoid overlap in subsequent plots
```
